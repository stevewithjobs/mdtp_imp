```python
import torch
import matplotlib.pyplot as plt

# 读取存储的两个 Tensor
loaded_tensor1 = torch.load('mask1.pt')
loaded_tensor2 = torch.load('mask2.pt')

# 将 1D Tensor 重塑为 (11 x 21) 的 2D 形状
reshaped_tensor1 = loaded_tensor1.view(11, 21).detach().cpu().numpy()
reshaped_tensor2 = loaded_tensor2.view(11, 21).detach().cpu().numpy()

count_greater_0_5_tensor1 = (loaded_tensor1 > 0.5).sum().item()
count_greater_0_5_tensor1 = (loaded_tensor2 > 0.5).sum().item()

print(count_greater_0_5_tensor1)
print(count_greater_0_5_tensor1)


# 创建一个图形，包含两个子图
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# 可视化第一个 Tensor
im1 = axs[0].imshow(reshaped_tensor1, cmap='viridis', aspect='auto')
fig.colorbar(im1, ax=axs[0])
axs[0].set_title('Mask 1 Visualization')
axs[0].set_xlabel('X Axis')
axs[0].set_ylabel('Y Axis')

# 可视化第二个 Tensor
im2 = axs[1].imshow(reshaped_tensor2, cmap='viridis', aspect='auto')
fig.colorbar(im2, ax=axs[1])
axs[1].set_title('Mask 2 Visualization')
axs[1].set_xlabel('X Axis')
axs[1].set_ylabel('Y Axis')

# 显示图形
plt.tight_layout()
plt.show()
```
